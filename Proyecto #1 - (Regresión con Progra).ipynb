```python
#ml.proyeccion(modelo,ml.validacion)
import numpy as np
proyeccion = ml.proyeccion(betas,ml.validacion)

proyeccion['SalePrice'] =ml.validacion.SalePrice
proyeccion['Error'] =  proyeccion['y_pred']-proyeccion['SalePrice']
proyeccion['% de error sobre la real'] = 100*np.abs(proyeccion['Error'])/proyeccion['SalePrice']
#display(proyeccion[3:6])
resumen = proyeccion.describe().round(2)
resumen['% de error sobre la real'] = np.round(resumen['% de error sobre la real'],2).astype(str)+'%'
display(resumen)
```
